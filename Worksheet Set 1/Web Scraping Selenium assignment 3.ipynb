```python
    #extracting the product name
    name_tag=driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]')
    
    name=[]
    for i in name_tag:
        if i.text is None:
            name.append('--')
        else:
            name.append(i.text)
        
    #extracting the price
    price_tag=driver.find_elements_by_xpath('//span[@class="a-price-whole"]')
    
    price=[]
    for i in price_tag:
        if i.text is None:
            price.append('--')
        else:
            price.append(i.text)
        
    #extracting delivery date
    expected_delivery_tag=driver.find_elements_by_xpath('//span[@class="a-color-base a-text-bold"]')
    
    expected_delivery=[]
    for i in expected_delivery_tag:
        if i.text is None:
            expected_delivery.append('--')
        else:
            expected_delivery.append(i.text)
   
    try:
        product_tag_url=driver.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
        
        product_url=[]
        for i in product_tag_url:
            if i.text is None:
                product_url.append('--')
            else:
                product_url.append(i.get_attribute('href'))
            
    except ElementNotVisibleException as e:
        print('Exception raised :',e)
    
    print(name)
    print(price)
    print(expected_delivery)
    print(product_url)
```
